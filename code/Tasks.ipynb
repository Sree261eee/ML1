```python
#task1

from keras.models import Sequential
from keras.layers.core import Dense, Activation

# load dataset
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

dataset = pd.read_csv("diabetes.csv", header=None).values

X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:8], dataset[:,8],test_size=0.25, random_state=87)
np.random.seed(155)
my_first_nn = Sequential() # create model
my_first_nn.add(Dense(20, input_dim=8, activation='relu'))#hidden layer
# #Adding more dense layers
my_first_nn.add(Dense(21, input_dim=8, activation='relu'))
my_first_nn.add(Dense(15, input_dim=8, activation='relu'))
my_first_nn.add(Dense(25, input_dim=8, activation='relu'))
my_first_nn.add(Dense(1, activation='sigmoid'))#output layer

my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100, initial_epoch=0)

print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))
```

```python
#task2 without preprocessing data
cancer_df = pd.read_csv("breastcancer.csv")
cancer_df["diagnosis"] = cancer_df["diagnosis"].map({"M":0, "B":1})
cancer_df.drop(columns=["id", "Unnamed: 32"], inplace=True)
cancer_df  = cancer_df.values
X_train_c, X_test_c, Y_train_c, Y_test_c = train_test_split(cancer_df[:,1:], cancer_df[:,0],test_size=0.25, random_state=87)
np.random.seed(155)
my_first_nnc = Sequential() # create model
my_first_nnc.add(Dense(20, input_dim=30, activation='relu'))#hidden layer
# #Adding more dense layers
my_first_nnc.add(Dense(21, input_dim=30, activation='relu'))
my_first_nnc.add(Dense(22, input_dim=30, activation='relu'))
my_first_nnc.add(Dense(23, input_dim=30, activation='relu'))
my_first_nnc.add(Dense(1, activation='sigmoid'))#output layer

my_first_nnc.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
my_first_nnc_fitted = my_first_nnc.fit(X_train_c, Y_train_c, epochs=100, initial_epoch=0)

print(my_first_nnc.summary())
print(my_first_nnc.evaluate(X_test_c, Y_test_c))
```

```python
#task3
from keras.models import Sequential         
from keras.layers.core import Dense

# load dataset
import pandas as pd
dataset = pd.read_csv("breastcancer.csv")    #loading dataset

dataset.drop(columns=["id", "Unnamed: 32"], inplace=True)
dataset["diagnosis"] = dataset["diagnosis"].map({"M":0, "B":1})
dataset
X = dataset.iloc[:, 1:].values
y = dataset.iloc[:, 0].values


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

#--------Data normalizes here---------#
from sklearn.preprocessing import StandardScaler     #applying standard scaling to normalize the data
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)                         #after scaling transforming on test data

my_first_nn = Sequential() # create model where sequential is linear stack of layers
my_first_nn.add(Dense(20, input_dim=30, activation='relu')) # hidden layer
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) #before training model configuring using learning process via compile method
my_first_nn_fitted = my_first_nn.fit(X_train, y_train, epochs=100, verbose=0, initial_epoch=0)   #training model

print("After Normalization performance resulyts are:\n")
print(my_first_nn.summary())                  #summary prints complete description
print(my_first_nn.evaluate(X_test, y_test))
```
